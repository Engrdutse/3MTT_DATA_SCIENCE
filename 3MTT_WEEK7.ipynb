**#Data Wrangling Techniques**

```python

# Importing necessary libraries
import pandas as pd
```

```python
url_data = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
```

```python
# Load the real dataset
titanic1 = pd.read_csv(url_data)


```

```python
# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(titanic1.head())

```

```python
# Check for missing values
print("\nMissing values in the dataset:")
print(titanic1.isnull().sum())

```

```python
# Dropping rows with missing values
titanic2 = titanic1.dropna()

```

```python
titanic2.head()
```

```python
# Dropping unnecessary columns
titanic2 = titanic2.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

```

```python
titanic2.head(5)
```

```python
# Convert categorical variables to dummy variables
titanic2 = pd.get_dummies(titanic2, columns=['Sex', 'Embarked'], drop_first=True)

```

```python
# Normalize numerical variables
titanic2['Age'] = (titanic2['Age'] - titanic2['Age'].mean()) / titanic2['Age'].std()
titanic2['Fare'] = (titanic2['Fare'] - titanic2['Fare'].mean()) / titanic2['Fare'].std()

```

```python
# Display the first 5 rows of the processed dataset
print("\nProcessed dataset:")
print(titanic2.head())

```

```python
# Save the processed dataset to a new CSV file
titanic2.to_csv('processed_titanic.csv', index=False)

print("\nDataset processing and wrangling is complete!")
```
