# Data Wrangling techniques using real-world dataset

# Importing necessary libraries


```python
#Impoting necesary libraries
import pandas as pd
```

```python
# Load the real dataset from a webseite

#url = "https://raw.github.com/Engrdutse/3MTT_DATA_SCIENCE/seaborn-data/main/tested.csv"

```

```python
titanic_data = pd.read_csv('tested.csv')
```

```python
# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
titanic_data.head()

```

```python
# Check for missing values
print("\nMissing values in the dataset:")
titanic_data.isnull().sum()

```

```python
# Dropping rows with missing values
titanic_data = titanic_data.dropna()

```

```python
titanic_data.head()
```

```python
# Dropping unnecessary columns
titanic_data = titanic_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
titanic_data.head()
```

```python
# Convert categorical variables to dummy variables
titanic_data = pd.get_dummies(titanic_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
titanic_data.head()
```

```python
# Normalize numerical variables
titanic_data['Age'] = (titanic_data['Age'] - titanic_data['Age'].mean()) / titanic_data['Age'].std()
titanic_data['Fare'] = (titanic_data['Fare'] - titanic_data['Fare'].mean()) / titanic_data['Fare'].std()

```

```python
# Display the first 5 rows of the processed dataset
print("\nProcessed dataset:")
print(titanic_data.head())
```

```python
# Save the processed dataset to a new CSV file
titanic_data.to_csv('processed_titanic.csv', index=False)

print("\nDataset processing and wrangling is complete!")
```
