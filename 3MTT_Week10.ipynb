```python
# Check for missig Values
titanic.isnull().sum()
```

```python
# We select Age column to perform the filling of the missing values
titanic.age.head(20)
```

```python
#Mean Inputation fo the missing value
mean_value = titanic['age'].mean()
print(mean_value)
```

```python
#We fill the missing age with mean age computed.
titanic['age'] = titanic['age'].fillna(mean_value)
```
