```python
#importing Nacessary Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
# Creating the Url link for the data
url = "https://raw.githubusercontent.com/Engrdutse/3MTT_DATA_SCIENCE/main/StudentsPerformance.csv"
```

```python
#Creating a dataframe from the url link
Students = pd.read_csv(url)
```

```python
#Displaying the dataset
Students.head()
```

```python
#Create a copy of the Data
stdP = Students.copy()
```

```python
stdP.head()
```

```python
#Displaying the coluumn levels
stdP.columns
```

```python

# Reanaming and shaping the column lebel with title Case
stdP = stdP.rename(columns={'race/ethnicity': 'Group'})

```

```python
stdP.head()
```

```python
stdP = stdP.rename(columns={'gender': 'Gender'})
```

```python
stdP = stdP.rename(columns={'parental level of education': 'Parental level of Education'})
```

```python
stdP.head()
```

```python
stdP = stdP.rename(columns={'math score': 'Math Score', 'reading score': 'Reading Score','writing score': 'Writing Score'})
```

```python
stdP.head(10) #Now the columns take Title Case
```

```python
#Removing unwanted columns from the data
stdP.drop(['lunch','test preparation course'], axis=1, inplace=True)
```

```python
stdP.head(10)
```

```python
#Checking for any missing value in the data
stdP.isnull().sum()
```
